```python

import pandas as pd
df = pd.read_csv("Superstore.csv" ,encoding='latin1')
print(df.head().to_string())
```

```python
# Shape and info
print("Rows & Columns:", df.shape)
print("\nColumn Names:\n", df.columns)

# Summary
print("\nSummary:\n", df.describe())

# Null values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Region-wise Sales
print("\nSales by Region:\n", df.groupby("Region")["Sales"].sum())

# Region-wise Profit
print("\nProfit by Region:\n", df.groupby("Region")["Profit"].sum())
```

```python
# Sales by Category
print("\nSales by Category:\n", df.groupby("Category")["Sales"].sum())

# Profit by Sub-Category
print("\nProfit by Sub-Category:\n", df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False))
```

```python
print("\nShipping Mode Impact on Profit:\n", df.groupby("Ship Mode")["Profit"].mean())
```

```python
import matplotlib.pyplot as plt
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(6,4))
region_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

##2
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

plt.figure(figsize=(8,6))
subcat_profit.plot(kind='barh', color='lightgreen')
plt.title('Profit by Sub-Category')
plt.xlabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()

##3

category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(5,5))
category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Category')
plt.ylabel('')  # removes y-axis label
plt.tight_layout()
plt.show()

##4

ship_profit = df.groupby('Ship Mode')['Profit'].sum()

plt.figure(figsize=(6,4))
ship_profit.plot(kind='bar', color='orange')
plt.title('Profit by Shipping Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```
